### iii) Replacing missing value using Mean along each column in Column 'Grade'

```python
##print out rows for columngrade with missing values
# Filter rows where 'Grade' column has null values
rows_nullgrade = df_dropna[df_dropna['Grade'].isnull()]

# Print the all the rows with missing values in column 'Grade'
rows_nullgrade
```

```python
# Replace missing values with the mean in a new DataFrame
mean_grade = df_dropna['Grade'].mean()

df_new_filled = df_dropna.copy()  # Create a copy 
df_new_filled['Grade'] = df_new_filled['Grade'].fillna(mean_grade)
```

```python
df_new_filled.info()
```

```python
##Check whether all the missing values in column grade is filled with the mean values for grade
mask = (df_new_filled['Grade'] == mean_grade)

# Print the rows where 'Grade' is equal to the mean
df_new_filled[mask]
```

# Step 6: Data Aggregation
* Group the data by a particular column and calculate summary statistics (e.g., mean, sum) for 
other columns within each group. For example, by grouping using groupby() the data by mont 
and calculating the total revenue for each mont.

#### i) Using groupby() for the column 'Location' to calculate summary statistics (e.g., mean, min, max) for other columns. This helps to provide insights into how housing prices, sizes, or other features vary across different regions within King County, USA.

```python
## Exclude columns with datatypes 'object'
numeric_col = df_step5.select_dtypes(include=['float64', 'int64'])

# Group by 'Location' and aggregate only numeric columns
summary_stats_location = df_step5.groupby('Location')[numeric_col.columns].agg(['min', 'max', 'mean'])

#Print the first 5th rows of the stats out
summary_stats_location.head()
```

#### ii) Using groupby() for the column 'Grade' to calculate summary statistics (e.g., mean, min, max) for specific columns like column 'Price','Bedrooms','Bathrooms','Grade','Condition'. This offer insights into how different grades impact on these 5 variables.

```python
# List of columns to include
columns_include = ['Price']

# Select only specified columns and filter numeric columns
numeric_col3 = df_step5[columns_include].select_dtypes(include=['float64', 'int64'])

# Group by 'Bedrooms' and aggregate only specified columns
summary_stats_bed = df_step5.groupby('Bedrooms')[numeric_col3.columns].agg(['min', 'max', 'mean'])

#Print the stats out
summary_stats_bed
```
